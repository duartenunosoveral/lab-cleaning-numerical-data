```python
from pandas.core.frame import DataFrame

def divide_numerical_categorical(dataframe: DataFrame) -> tuple[DataFrame, DataFrame]:
    """
    will divide numerical and categorical columns of the input DataFrame.

    Parameters:
        dataframe (pandas.core.frame.DataFrame): The input DataFrame.

    Returns:
        tuple[pandas.core.frame.DataFrame, pandas.core.frame.DataFrame]: 
        A tuple containing two DataFrames:
            - The first DataFrame contains the numerical columns.
            - The second DataFrame contains the categorical columns.
    """
    numerical_cols = dataframe.select_dtypes(include='number')
    categorical_cols = dataframe.select_dtypes(exclude='number')

    return numerical_cols, categorical_cols
```

```python
def separate_continuous_discrete(dataframe: pd.DataFrame, unique_threshold: int = 36) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Separates numerical columns of the input DataFrame into continuous and discrete columns.

    Parameters:
        dataframe (pandas.core.frame.DataFrame): The input DataFrame.
        unique_threshold (int, optional): Threshold to discriminate between continuous and discrete columns
            based on the number of unique values. Default is 36.

    Returns:
        tuple[pandas.core.frame.DataFrame, pandas.core.frame.DataFrame]: 
        A tuple containing two DataFrames:
            - The first DataFrame contains the numerical continuous columns.
            - The second DataFrame contains the numerical discrete columns.
    """
    continuous_cols = []
    discrete_cols = []

    for column in dataframe.select_dtypes(include='number'):
        unique_values = dataframe[column].nunique()
        if unique_values > unique_threshold:
            continuous_cols.append(column)
        else:
            discrete_cols.append(column)

    continuous_df = dataframe[continuous_cols]
    discrete_df = dataframe[discrete_cols]

    return continuous_df, discrete_df
```

```python
continuous_df, discrete_df = separate_continuous_discrete(customer_df)
```

```python
def create_barplots(discrete_df, figsize=(16,16), plot_func=sns.countplot):
    """
    Creates bar plots for each column in the discrete_df DataFrame using the specified plotting function.

    Parameters:
        discrete_df (pandas.core.frame.DataFrame): The DataFrame containing discrete columns.
        figsize (tuple, optional): Figure size. Default is (16, 16).
        plot_func (function, optional): Plotting function to create bar plots. Default is sns.countplot.
    """
  
    plt.figure(figsize=figsize)

    for column in discrete_df.columns:
        plt.subplot(len(discrete_df.columns), 1, discrete_df.columns.get_loc(column) + 1)
        plot_func(x=column, data=discrete_df)
        plt.xlabel(column)
        plt.ylabel('Count')

    plt.tight_layout()

    plt.show()
```

```python
create_barplots(discrete_df, figsize=(16,16), plot_func=sns.barplot)
```

```python
def create_histograms(continuous_df, figsize=(16, 16), plot_func=sns.histplot):
    """
    Creates histograms for each column in the continuous_df DataFrame using the specified plotting function.

    Parameters:
        continuous_df (pandas.core.frame.DataFrame): The DataFrame containing continuous columns.
        figsize (tuple, optional): Figure size. Default is (16, 16).
        plot_func (function, optional): Plotting function to create histograms. Default is sns.histplot.
    """
    plt.figure(figsize=figsize)

    for column in continuous_df.columns:
        plt.subplot(len(continuous_df.columns), 1, continuous_df.columns.get_loc(column) + 1)
        plot_func(data=continuous_df, x=column, kde=True)
        plt.xlabel(column)
        plt.ylabel('Count')
        
    plt.tight_layout()

    plt.show()
```

```python
create_histograms(continuous_df, figsize=(16,16), plot_func=sns.histplot)
```

```python
def create_boxplots(continuous_df, figsize=(10, 8)):
    """
    Creates box plots for each column in the continuous_df DataFrame.

    Parameters:
        continuous_df (pandas.core.frame.DataFrame): The DataFrame containing continuous columns.
        figsize (tuple, optional): Figure size. Default is (10, 8).
    """

    plt.figure(figsize=figsize)

    for column in continuous_df.columns:
        plt.subplot(len(continuous_df.columns), 1, continuous_df.columns.get_loc(column) + 1)
        sns.boxplot(x=column, data=continuous_df)
        plt.xlabel(column)

    plt.tight_layout()

    plt.show()
```

```python
create_boxplots(continuous_df, figsize=(16, 16))
```
